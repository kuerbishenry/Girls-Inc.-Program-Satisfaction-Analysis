```python
combined = pd.concat([Eur_Agg, Aft_Agg, Well_Agg])
combined = combined.drop_duplicates(subset = ["Question"], keep = "first")
combined
```

```python
topic_map = {
    "Academic Interest": [
        "[I get excited about science.]",
        "[I am curious to learn more about science, computers, or technology.]",
        "[I am good at math.]",
        "[I know what education or training I will need for the career I want.]",
        "[I care about doing well in school.]",
        "[If something interests me, I try to learn more.]"
    ],
    "Safety": [
        "[There are adults at Girls Inc. I can depend on to help me.]",
        "[Adults at Girls Inc. listen to what I have to say.]",
        "[I feel safe at Girls Inc.]",
        "[There are adults at Girls Inc. who help me think about my future.]"
    ],
    "Emotional Regulation": [
        "[When I get angry or frustrated, I lose my temper.]",
        "[I lose control when I do not get my way.]",
        "[When a friend says something mean to me, I get over it quickly.]",
        "[I do my best even when things get tough.]",
        "[I can control the important things in my life.]"
    ],
    "Confidence": [
        "[I am good at leading others to reach a goal.]",
        "[Girls like me can be leaders.]",
        "[At Girls Inc., I get the chance to be a leader.]",
        "[At Girls Inc., girls learn they can do anything boys can do.]",
        "[At Girls Inc., I get to try new things.]",
        "[I know I can handle my problems.]",
        "[I am happy with how my body looks.]"
    ],
    "Opurtunities": [
        "[I know what education or training I will need for the career I want.]",
        "[I know adults I can talk to who work in a career that interests me.]",
        "[I know about different financial aid options for college.]",
        "[At Girls Inc., girls work together on projects and activities.]"
    ],
    "Diversity": [
        "[I get along well with people of different races, cultures, and religions.]",
        "[I get along well w/ people of different races, cultures, and religions]",
        "[I listen to other people’s ideas and feelings.]",
        "[I listen to people even if I disagree with them.]",
        "[At Girls Inc., I get to learn more about people who are different from me.]",
        "[I get to learn more about people who are different from me.]"
    ]
}

combined["Question"] = combined["Question"].astype(str).str.strip()

topic_map_clean = {}
for category, questions in topic_map.items():
    for q in questions:
        topic_map_clean[q.strip()] = category

combined["Category"] = combined["Question"].map(topic_map_clean)
```

```python
combined
```

```python
response_cols = [
    "Strongly Agree", "Agree", "Disagree", "Strongly Disagree",
    "All the time", "Most of the time", "Some of the time", "None of the time"
]

rows = []

for _, row in combined.iterrows():
    total = sum(row.get(col, 0) for col in response_cols)

    if total == 0 or pd.isna(row.get("Category")):
        continue

    if row.get("Strongly Agree", 0) + row.get("Agree", 0) + row.get("Disagree", 0) + row.get("Strongly Disagree", 0) > 0:
        strong_positive = row.get("Strongly Agree", 0) / total * 100
        positive = row.get("Agree", 0) / total * 100
        negative = row.get("Disagree", 0) / total * 100
        strongly_negative = row.get("Strongly Disagree", 0) / total * 100
    else:       
        strong_positive = row.get("All the time", 0) / total * 100
        positive = row.get("Most of the time", 0) / total * 100
        negative = row.get("Some of the time", 0) / total * 100
        strongly_negative = row.get("None of the time", 0) / total * 100

    rows.append({
        "Category": row["Category"],
        "strongly positive": strong_positive,
        "positive": positive,
        "negative": negative,
        "strongly negative": strongly_negative
    })

percent_df = pd.DataFrame(rows)

pct_cols = ["strongly positive", "positive", "negative", "strongly negative"]

plot_df = percent_df.groupby("Category")[pct_cols].mean().reindex([
    "Academic Interest",
    "Safety",
    "Emotional Regulation",
    "Confidence",
    "Opurtunities",
    "Diversity"
])

colors = ["green", "yellow", "orange", "red"]

x = np.arange(len(plot_df.index))
bottom = np.zeros(len(plot_df))

plt.figure(figsize=(14, 10))

for i, col in enumerate(pct_cols):
    plt.bar(
        x,
        plot_df[col].fillna(0).values,
        bottom=bottom,
        color=colors[i],
        label=col
    )
    bottom += plot_df[col].fillna(0).values

plt.xticks(x, plot_df.index, rotation=45, ha="right", fontsize=12)
plt.ylabel("Percentage")
plt.xlabel("Category")
plt.title("Positive and Negative Response Percentages by Category")
plt.legend(title="Response Type")
plt.tight_layout()
plt.show()
```

```python
for category in [
    "Academic Interest",
    "Safety",
    "Emotional Regulation",
    "Confidence",
    "Opurtunities",
    "Diversity"
]:
    print(f"\n{category}:")
    cat_row = plot_df.loc[category]

    for col in pct_cols:
        print(f"  {col}: {cat_row[col]:.2f}%")
```
